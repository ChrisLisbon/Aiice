## Data Preparation

Let's get data for past 5 years and split them:
- 2025 year - test data
- 2024 year - val data
- 2020-2023 - train data

Also, we're going to get date per 7 days in order to simplify our example.
Our model will predict next 4 days by last 12 days per 7 days (next 4 weeks by last 12 weeks)

```python
start = datetime(2020, 1, 1)
step = 7
pre_history_len = 12
forecast_len = 4
```

```python
step = 7
start = datetime(2020, 1, 1)

loader = Loader()
data = loader.get(start=start.date(), step=step, tensor_out=True)
```

```python
data.shape
```

Also let's reduce our tensors with `apply_downsample` in order to minimize compute time of out future model

```python
# delete each 2nd pixel by 1 and 2 axes
downsampled_date = apply_downsample(data, i=2, axes=(1, 2))
downsampled_date.shape
```

For now we have 287 days, with the code below we can split them:

```python
train_idx, val_idx, test_idx = [], [], []

for i in range(len(downsampled_date)):
    current_date = start + timedelta(days=i*step)
    year = current_date.year
    
    if 2020 <= year <= 2023:
        train_idx.append(i)
    elif year == 2024:
        val_idx.append(i)
    elif year == 2025:
        test_idx.append(i)

train = downsampled_date[train_idx]
val = downsampled_date[val_idx]
test = downsampled_date[test_idx]

print(train.shape, val.shape, test.shape)
```
